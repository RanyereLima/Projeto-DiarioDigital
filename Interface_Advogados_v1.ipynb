```python
import os
processos = [{'numero': '12345', 'cliente': 'JoÃ£o Silva', 'natureza': 'Trabalhista', 'ativo': True},
             {'numero': '67890', 'cliente': 'Maria Oliveira', 'natureza': 'Civil', 'ativo': False}]

def exibir_nome_do_programa():
    '''Exibe o nome estilizado do programa na tela'''
    print("""ğ•€ğ•Ÿğ•¥ğ•–ğ•£ğ•—ğ•’ğ•”ğ•– ğ”¸ğ••ğ•§ğ• ğ•˜ğ•’ğ••ğ• ğ•¤""")

def exibir_opcoes():
    '''Exibe as opÃ§Ãµes disponÃ­veis no menu principal'''
    print('1. Cadastrar processo')
    print('2. Listar processos')
    print('3. Alternar estado do processo')
    print('4. Sair\n')

def finalizar_app():
    '''Exibe a mensagem de finalizaÃ§Ã£o do app'''
    exibir_subtitulo('Finalizando o app...')

def voltar_menu_principal():
    ''' Solicita uma tecla para voltar ao menu principal '''
    input('\nDigite uma tecla para voltar ao menu inicial ')
    main()

def exibir_subtitulo(texto):
    ''' Exibe um subtÃ­tulo estilizado na tela '''
    os.system('cls')
    linha = '*' * len(texto)
    print(linha)
    print(texto)
    print(linha)
    print()

def cadastrar_novo_processo():
    '''FunÃ§Ã£o para cadastrar novos processos'''
    exibir_subtitulo('Cadastro de novos processos')
    numero = input('Digite o nÃºmero do processo: ')
    cliente = input('Digite o nome do cliente: ')
    natureza = input('Digite a natureza do processo (Civil, Trabalhista, FamÃ­lia, Comercial): ')
    dados_do_processo = {'numero': numero, 'cliente': cliente, 'natureza': natureza, 'ativo': True}
    processos.append(dados_do_processo)
    print(f'O processo {numero} foi cadastrado com sucesso!')
    voltar_menu_principal()

def listar_processos():
    '''Lista os processos cadastrados'''
    exibir_subtitulo('Listando os processos')
    print(f'{{NÃºmero.ljust(10)}} | {{Cliente.ljust(20)}} | {{Natureza.ljust(15)}} | Status')
    for processo in processos:
        numero = processo['numero']
        cliente = processo['cliente']
        natureza = processo['natureza']
        ativo = 'Ativo' if processo['ativo'] else 'Inativo'
        print(f'{numero.ljust(10)} | {cliente.ljust(20)} | {natureza.ljust(15)} | {ativo}')
    voltar_menu_principal()

def alternar_estado_processo():
    '''Altera o estado ativo/inativo de um processo'''
    exibir_subtitulo('Alternando estado do processo')
    numero_processo = input('Digite o nÃºmero do processo que deseja alterar o estado: ')
    processo_encontrado = False
    for processo in processos:
        if numero_processo == processo['numero']:
            processo_encontrado = True
            processo['ativo'] = not processo['ativo']
            mensagem = f'O processo {numero_processo} foi ativado com sucesso!' if processo['ativo'] else f'O processo {numero_processo} foi desativado com sucesso!'
            print(mensagem)
    if not processo_encontrado:
        print('O processo nÃ£o foi encontrado')
    voltar_menu_principal()

def opcao_invalida():
    ''' Exibe mensagem de opÃ§Ã£o invÃ¡lida e retorna ao menu principal '''
    print('OpÃ§Ã£o invÃ¡lida\n')
    voltar_menu_principal()

def escolher_opcao():
    '''Solicita e executa a opÃ§Ã£o escolhida pelo usuÃ¡rio'''
    try:
        opcao_escolhida = int(input('Escolha uma opÃ§Ã£o: '))

        if opcao_escolhida == 1:
            cadastrar_novo_processo()
        elif opcao_escolhida == 2:
            listar_processos()
        elif opcao_escolhida == 3:
            alternar_estado_processo()
        elif opcao_escolhida == 4:
            finalizar_app()
        else:
            opcao_invalida()
    except ValueError:
        opcao_invalida()

def main():
    '''FunÃ§Ã£o principal que inicia o programa'''
    os.system('cls')
    exibir_nome_do_programa()
    exibir_opcoes()
    escolher_opcao()

if __name__ == '__main__':
    main()
```
